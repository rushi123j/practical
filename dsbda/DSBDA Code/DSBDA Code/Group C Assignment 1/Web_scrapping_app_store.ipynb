# **Web Scraping Amazon Reviews with Python and Beautiful Soup**

Web scraping is a technique used by programmers to automate the process of finding and extracting data from websites within a short period of time. It can be difficult to get large amounts of data using traditional data collection methods. We can automate the process of capturing data from websites by using web scraping tools. Web scraping allows us to collect unstructured data and store it in a structured format where it can be used for further analysis.

Web scraping can be used for a wide variety of purposes including:

Price Comparison: used to collect data from online shopping websites and compare products and pricing.

Real Estate: used to collect and build databases for real estate listings for market analysis.

Social Media: used to collect data from Social Media websites to find out what topics are trending.

Lead Generation: used to collect contact information including names, phone numbers, and email addresses to communicate with potential customers.

Movie Reviews: used to collect information about movies to help decide which movies are popular and highly rated.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urlencode
import csv
```

```python
# Define a list of URL's that will be scraped.
list_of_urls = ['https://www.amazon.com/Apple-Watch-Cellular-Midnight-Aluminum/product-reviews/B09PB1PTH8/ref=cm_cr_othr_d_show_all_btm?ie=UTF8&reviewerType=all_reviews',
                'https://www.amazon.com/Apple-Watch-Cellular-Midnight-Aluminum/product-reviews/B09PB1PTH8/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber=2',
                'https://www.amazon.com/Apple-Watch-Cellular-Midnight-Aluminum/product-reviews/B09PB1PTH8/ref=cm_cr_getr_d_paging_btm_next_3?ie=UTF8&reviewerType=all_reviews&pageNumber=3']
                
```

```python
# Retrieve each of the url's HTML data and convert the data into a beautiful soup object.  
# Find, extract and store reviewer names and review text into a list.

names = []
reviews = []
data_string = ""

for url in list_of_urls: 
    params = {'api_key': "29d44f059aef7fcdeaf843f2a707557c", 'url': url}
    response = requests.get('http://api.scraperapi.com/',   params=urlencode(params))
    soup = BeautifulSoup(response.text, 'html.parser')

    for item in soup.find_all("span", class_="a-profile-name"):
      data_string = data_string + item.get_text()
      names.append(data_string)
      data_string = ""  
    
    for item in soup.find_all("span", {"data-hook": "review-body"}):
      data_string = data_string + item.get_text()
      reviews.append(data_string)
      data_string = ""
```

```python
# Create the dictionary.
reviews_dict = {'Reviewer Name': names, 'Reviews': reviews}
# Print the lengths of each list.  
print(len(names), len(reviews))
```

```python
# Create a new dataframe.
df = pd.DataFrame.from_dict(reviews_dict, orient='index')
df.head()
```

```python
# Delete all the columns that have missing values.
df.dropna(axis=1, inplace=True)
df.head()
```

```python
# Transpose the dataframe.
prod_reviews = df.T
print(prod_reviews.head(10))
```

```python
# Remove special characters from review text.
prod_reviews['Reviews'] = prod_reviews['Reviews'].str.replace('\n','')
prod_reviews.head(5)
```

```python
# Convert dataframe to CSV file.
prod_reviews.to_csv('reviews.csv', index=False, header=True)
```
